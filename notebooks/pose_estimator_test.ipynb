```python
import sys
sys.path.append("../src")

import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import mediapipe as mp
import time

from rehab_ai.pose_estimator import PoseEstimator
from rehab_ai.pose_analyzer import PoseAnalyzer
from config.settings import ModelTypePath
```

```python
pose_estimator = PoseEstimator(model_path=ModelTypePath.HEAVY)
```

```python
print(mp.tasks.vision.PoseLandmark.LEFT_SHOULDER)
from matplotlib.animation import FuncAnimation
import plotly.graph_objects as go
```

```python
%matplotlib inline
%matplotlib widget

stream = cv.VideoCapture(0) # 0 = default webcam available
if not stream.isOpened():
    raise RuntimeError("Could not open live stream.")

shoulder_x = []
shoulder_y = []
shoulder_z = []

elbow_x = []
elbow_y = []
elbow_z = []

wrist_x = []
wrist_y = []
wrist_z = []

while True:

    success, frame = stream.read()

    if not success:
        break

    # Convert BGR to RGB
    rgb_frame = cv.cvtColor(frame, cv.COLOR_BGR2RGB)

    # Convert to MediaPipe image
    mp_image = mp.Image(
        image_format=mp.ImageFormat.SRGB,
        data=rgb_frame
    )

    # Timestamp (milliseconds)
    timestamp_ms = int(time.time() * 1000)

    pose = pose_estimator.detect_pose(mp_image, timestamp_ms)
    angles = PoseAnalyzer(pose)
    
    if pose is not None and pose.is_detected:
        annotated_image = pose.draw_landmarks(frame)
        # shoulder_x.append(pose.get_landmark(mp.tasks.vision.PoseLandmark.RIGHT_SHOULDER).x)
        # shoulder_y.append(pose.get_landmark(mp.tasks.vision.PoseLandmark.RIGHT_SHOULDER).y)
        # shoulder_z.append(pose.get_landmark(mp.tasks.vision.PoseLandmark.RIGHT_SHOULDER).z)

        # elbow_x.append(pose.get_landmark(mp.tasks.vision.PoseLandmark.RIGHT_ELBOW).x)
        # elbow_y.append(pose.get_landmark(mp.tasks.vision.PoseLandmark.RIGHT_ELBOW).y)
        # elbow_z.append(pose.get_landmark(mp.tasks.vision.PoseLandmark.RIGHT_ELBOW).z)

        # wrist_x.append(pose.get_landmark(mp.tasks.vision.PoseLandmark.RIGHT_WRIST).x)
        # wrist_y.append(pose.get_landmark(mp.tasks.vision.PoseLandmark.RIGHT_WRIST).y)
        # wrist_z.append(pose.get_landmark(mp.tasks.vision.PoseLandmark.RIGHT_WRIST).z)
        print(f"{angles.right_shoulder} degrees")

    else:
        annotated_image = frame
    cv.imshow("Pose Estimation", annotated_image)
    if cv.waitKey(1) == ord('q'):
        pose_estimator.close()
        stream.release()
        cv.destroyAllWindows()
        break


```
